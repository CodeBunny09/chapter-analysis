```python
import nltk
from nltk import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from collections import Counter
```

```python
# Reading and tokenization
text = open("master.txt", "r", encoding = "UTF-8")

raw = text.read()
tokens = word_tokenize(raw)

word_count = Counter(tokens)
word_count.most_common(10)
```

```python
# Normalization
tokens = [word for word in tokens if word.isalpha()]
tokens = [word.lower() for word in tokens]


word_count = Counter(tokens)
word_count.most_common(10)
```

```python
# Removing stop words
tokens = [word for word in tokens if not word in stopwords.words("english")]

word_count = Counter(tokens)
word_count.most_common(10)
```

```python
# Stemming
from nltk.stem import PorterStemmer, SnowballStemmer 

snowball = SnowballStemmer("english")

s_tokens = [snowball.stem(token) for token in tokens]

s_word_count = Counter(s_tokens)
s_word_count.most_common(10)
```

```python
# Lemmatization
from nltk import WordNetLemmatizer

lemma = WordNetLemmatizer()

l_tokens = [lemma.lemmatize(token, pos = "a") for token in tokens] #For Adjectives
l_tokens = [lemma.lemmatize(token, pos = "s") for token in tokens] #For Sattelite Adjectives
l_tokens = [lemma.lemmatize(token, pos = "r") for token in tokens] #For Adverbs
l_tokens = [lemma.lemmatize(token, pos = "n") for token in tokens] #For Nouns
l_tokens = [lemma.lemmatize(token, pos = "v") for token in tokens] #For Verbs

l_word_count = Counter(l_tokens)
l_word_count.most_common(10)
```

```python
# Creating a WordCloud

from wordcloud import WordCloud
import matplotlib.pyplot as plt

word_string = ""
for token in tokens:
    word_string += token + " "

wc = WordCloud(width=400, height=400, collocations = False, background_color = 'black', max_words=200)
cloud = wc.generate(word_string)

plt.figure(figsize=(200, 200))
plt.clf()
plt.imshow(cloud)
plt.axis('off')
plt.title("Motivation WordCloud")
plt.savefig('wordcloud.png')
plt.show()
```

```python
# Summing ot up in a function

import nltk
from nltk import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from collections import Counter

from nltk import WordNetLemmatizer

from wordcloud import WordCloud
import matplotlib.pyplot as plt

def create_cloud(source="./master.txt", img_name="WordCloud.png"):
    # Reading and tokenization
    text = open(source, "r", encoding = "UTF-8")

    raw = text.read()
    tokens = word_tokenize(raw)

    # Normalization
    tokens = [word for word in tokens if word.isalpha()]
    tokens = [word.lower() for word in tokens]

    # Removing stop words
    tokens = [word for word in tokens if not word in stopwords.words("english")]

    # Lemmatization

    lemma = WordNetLemmatizer()

    tokens = [lemma.lemmatize(token, pos = "a") for token in tokens] #For Adjectives
    tokens = [lemma.lemmatize(token, pos = "s") for token in tokens] #For Sattelite Adjectives
    tokens = [lemma.lemmatize(token, pos = "r") for token in tokens] #For Adverbs
    tokens = [lemma.lemmatize(token, pos = "n") for token in tokens] #For Nouns
    tokens = [lemma.lemmatize(token, pos = "v") for token in tokens] #For Verbs

    word_count = Counter(tokens)
    print(word_count.most_common(10))

    # Creating the Word Cloud
    word_string = ""
    for token in tokens:
        word_string += token + " "

    wc = WordCloud(width=400, height=400, collocations = False, background_color = 'black', max_words=200)
    cloud = wc.generate(word_string)

    plt.figure(figsize=(400, 400))
    plt.clf()
    plt.imshow(cloud)
    plt.axis('off')
    plt.title("WordCloud")
    plt.savefig(img_name)
    plt.show()
    
    return word_count
```
